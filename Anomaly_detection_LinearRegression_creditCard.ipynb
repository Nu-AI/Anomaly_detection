```python
#distribution of classes (legit (=class 0) and fraudulent transaction (=class 1) ) to find if the data is imbalanced or not

#the row class gives the info about it
df_raw['Class'].value_counts()
```

we can see that class 0 has 284315 data while class 1 has only 492 data which is imblanced. more than 99% belongs to class 0. Therefore training with this datset gives a biased output (always favors for class 0)

```python
#seaprating legit and fraudulent classes

legit=df_raw[df_raw.Class==0]
fraud=df_raw[df_raw.Class==1]

#shapes of two classes
print(legit.shape)

print(fraud.shape)
```

```python
#statistical information about the amount spent in USD
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare the mean values for both classes in every features as it provides the importnat info about legit or faudulent class
df_raw.groupby('Class').mean()
```

```python
# dealing with imbalnced data by undersampling (deleting samples from the majority case)
#build a sample data set containing similar distribution of normal transaction and the fraudulent transaction
#number of fraudulent transaction =492. Randomly sample class 0 with 492 and make a data set containing 492 class 0 and 492 class 1 data samples

legit_sample=legit.sample(n=492)
print(legit_sample.shape)
```

```python
#creating a new dataset containing equal number of classes 

#concatenate legit_sample and fraud

df_new=pd.concat([legit_sample, fraud], axis =0) 

# axis =0 adds row wise . that is df_new contains first 492 data from legit_sample (class 0) and the next 492 from fraus(class 1)
print(df_new.shape)
```

```python
df_new.tail()
```

```python
#to check the statistical mean of different features for 2 classes
df_new.groupby('Class').mean()
```

comparing the mean values from df_raw and df_new, we can see that mean value has not changed so much. so the distribuition is same

```python
# splitting the data into input features and output targets
#input features contains time, v1,..V28 and target contains class 

X=df_new.drop(columns= 'Class',axis=1)
print(X.shape)
Y=df_new['Class']
print(Y.shape)
```

```python
#Splitting into training data and test data

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
#Model Training using logistic regression

model = LogisticRegression()

# training the Logistic Regression Model with Training Data
model.fit(X_train, Y_train)
```

```python
# Model Prediction

#to find the accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)

print("Accuracy of training: ",training_data_accuracy )
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print("Accuracy of testing: ",test_data_accuracy )
```
